```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df=pd.read_csv('CLEAN STORE.csv',low_memory=False)
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

BUILDING CENTER

```python
basket = (df[df['Store Name'] =="BUSY BEAVER BUILDING CENTERS"]
          .groupby(['City', 'Item Description'])['Net Sales Units']
          .sum().unstack().reset_index().fillna(0)
          .set_index('City'))
```

```python
def encode(x):
    if x > 0:
        return 1
    else:
        return 0
```

```python
basket=basket.applymap(encode)
```

```python
basket
```

```python
frequent=apriori(basket,min_support=0.06,use_colnames=True)
```

```python
frequent=pd.DataFrame(frequent)
```

```python
frequent
```

```python
rules=association_rules(frequent,metric='lift',min_threshold=1)
```

```python
rules.to_csv('BUSY BEAVER BUILDING CENTER.csv')
```
